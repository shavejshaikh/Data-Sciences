```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
df=pd.read_csv('http://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

### Univariate

```python
df_setosa   =df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versica  =df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']))
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']))
plt.plot(df_versica['sepal_length'],np.zeros_like(df_versica['sepal_length']))
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
```
