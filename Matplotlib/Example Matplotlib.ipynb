# Axes Labels ,Legend,Grid


```python
days =[1,2,3,4,5,6,7]
max_t=[32,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best",shadow="True",fontsize='small')#IMport to set the label 

plt.grid()

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
```

# Bar Chart

```python
import numpy as np
```

```python
company=['Google','Amazon','Microsoft','Facebook']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.bar(company,revenue,label="Revenue")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Billionaire Chart")
plt.bar(company,profit,label='Profit')
plt.legend(loc='best')
```

```python
plt.barh(company,revenue,label="Revenue")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Billionaire Chart")
plt.barh(company,profit,label='Profit')
plt.legend(loc='best')
```

# Histogram


```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g',histtype='step')
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_wom=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_sugar_men,blood_sugar_wom],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'])
plt.legend()

plt.xlabel("Sugar Range")
plt.ylabel("Total number of Patient")
plt.title("Sugar Level")
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_wom=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_sugar_men,blood_sugar_wom],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()

plt.xlabel("Sugar Range")
plt.ylabel("Total number of Patient")
plt.title("Sugar Level")
```

# Pie Chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']

plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0.1,0.2,0,0],startangle=45)
plt.show()
```
